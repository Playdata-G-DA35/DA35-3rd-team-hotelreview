```python
import os
import re
import nltk
import konlpy
from konlpy.tag import Okt

# 저장된 리뷰 데이터 가져오기
with open("hotel_review.txt", "r", encoding='utf-8') as f:
    review = f.read()

# 'n번째 리뷰.'를 기준으로  
tok_review = re.split(r"\d+번째 리뷰.", review)
# tok_review = nltk.word_tokenize(tok_review)
tok_words = []
for r in tok_review:
    tok_r = nltk.word_tokenize(r)
    tok_words.extend(tok_r)    
```

```python
okt = Okt()

okt_words = []
for word in tok_words:
    okt_words.extend(okt.pos(word, stem=True))
okt_words = [token for token, pos in okt_words if pos in [ 'Verb']]
```

```python
from nltk import Text
from wordcloud import WordCloud

okt_w = Text(okt_words, name='토큰화')
okt_w = okt_w.vocab()
wc = WordCloud(
    font_path=r"C:\\Windows\\Fonts\\malgunbd.ttf",
    max_words=100,
    min_font_size=1,
    max_font_size=50,
    relative_scaling=0.8,
    background_color="white"
)

word_cloud_img = wc.generate_from_frequencies(okt_w)
word_cloud_img.to_file("words_vocab.png")
```
